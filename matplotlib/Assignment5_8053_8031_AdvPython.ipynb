```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("Population.csv")
df
```

```python
df["Rural Population"]=df["Population"]-df["Urban Population"]
df
```

```python
plt.plot(df["Year"], df["Rural Population"], color = "#F54927", lw=0.8, ls='--', label='Rural Population')
plt.plot(df["Year"], df["Urban Population"], color = "#27F5E0", lw=0.8, ls='-.', label='Urban Population of India')
plt.xlabel('Year')
plt.ylabel('rural Population')
plt.legend()
plt.title('urban and rural population of India over the years')
plt.show()
```

```python
# Using a Bar chart display India's total and rural population from year 2020 to 2025
```

```python
import matplotlib.pyplot as plt
import pandas as pd
filtered_df = df[df['Year'] >= 2020]
x = filtered_df['Year']
fig,ax = plt.subplots(figsize = (10,6))
width = 0.39
plt.bar(x - width/2, filtered_df['Population'], width = width, label='Total Population', color='#27F5E0')
plt.bar(x + width/2, filtered_df['Rural Population'], width = width, label='Rural Population', color='#F54927')
plt.xlabel('Year')
plt.ylabel('Population (in Billions)')
plt.title("India's Total and Rural Population (2020 - 2025)")
plt.xticks(x,) 
plt.legend()
plt.show()
```

```python
# Using Histogram show distribution of developers by their experience [use survey_data_sample.csv]
```

```python
df = pd.read_csv("survey_results-sample.csv")
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')

dev_exp = df['YearsCode'].dropna()

plt.figure(figsize=(10, 6))

plt.hist(dev_exp, bins=20, color='#FFA6C9', edgecolor='black', alpha=0.7)

plt.title('Distribution of Developers by Years of Experience')
plt.xlabel('Years of Coding Experience')
plt.ylabel('Number of Developers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
