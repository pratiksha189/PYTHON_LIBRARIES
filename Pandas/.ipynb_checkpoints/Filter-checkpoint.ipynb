```python
import pandas as pd
```

```python
df = pd.read_csv('survey_results.csv')
df
```

```python
df.columns
```

```python
filt = df['ConvertedCompYearly'] > 1000000
df.loc[filt, ['Age', 'Country', 'YearsCode', 'LanguageHaveWorkedWith', 'Employment']]
```

```python
filt = df['Country'].isin(['United States of America', 'India', 'Japan'])
df.loc[filt, ['Age', 'Country', 'YearsCode', 'LanguageHaveWorkedWith', 'Employment']]
```

```python
filt = (df['ConvertedCompYearly'] > 1000000) & (df['Country'].isin(['United States of America', 'India', 'Japan']))
df_filtered=df.loc[filt, ['Age', 'Country', 'YearsCode', 'LanguageHaveWorkedWith', 'Employment']]
df_filtered
```

```python
df_filtered.info()
```

```python
filt = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
df.loc[filt, ['Age', 'Country', 'YearsCode', 'LanguageHaveWorkedWith', 'Employment']]
```
