```python
# Q.3Read csv gapminder2007.csv
# a.	Display data for all Asian countries
# b.	Display top and bottom 15 rows
# c.	Show all the rows where life expectancy is more than 50 years and less than 80 years
# d.	Show data for India, America and France
# e.	Show data for all countries where name starts with ‘A’
# f.	Show min/max and mean GDP and Population for each continent
# g.	Show population and life expectancy for all countries 
# h.	Sort the data by country names (ascending) and GDP (descending)
# i.	Find top 20 most populous countries
# j.	Delete all rows for a specific country
# k.	Change all column names to title case
# l.	Change column names: pop to Population, gdpPercap to GDP_Percap
# m.	Increase Life Expectancy by 2 years for all African countries
# n.	Find all the rows that contain NA values
# o.	Fill NA values with 0 for numeric columns
```

```python
df=pd.read_csv("gapminder2007.csv")
df
```

```python
# Display data for all Asian countries
df[df["continent"] == "Asia"]
```

```python
# Display top and bottom 15 rows
df.head(15)
```

```python
# Display top and bottom 15 rows
df.tail(15)
```

```python
# Show all the rows where life expectancy is more than 50 years and less than 80 years
df[(df["lifeExp"]>50) & (df["lifeExp"]<80 )]
```

```python
# Show data for India, America and France
```

```python
df[df["country"].isin(["India","America","France"])]
```

```python
# Show data for all countries where name starts with ‘A’
```

```python
df[df["country"].str.startswith('A')]
```

```python
# Show min/max and mean GDP and Population for each continent
df.groupby('continent').agg({
    'gdpPercap': ['min','max','mean'],
    'pop' : ['min','max','mean']
})
```

```python
# Show population and life expectancy for all countries 
df[['country','pop','lifeExp']]
```

```python
# Sort the data by country names (ascending) and GDP (descending)
df.sort_values(by = ['country','gdpPercap'],ascending=[True,False])
```

```python
df.sort_values(by = ['pop'],ascending=False)
```

```python
df.tail(20)
```

```python
# Delete all rows for a specific country
df = df[df['Country'] != 'Sweden']
df
```

```python
# Change all column names to title case
df.columns = [col.title() for col in df.columns]
df.columns
```

```python
# Change specific column names (Note: names reflect title case from previous case)
df = df.rename(columns={'Gdppercap': 'GDP_Percap'})
df
```

```python
# Increase Life Expectancy by 2 years for all African countries
# (Note: Using .loc to avoid SettingWithCopy warnings)
df.loc[df['Continent'] == 'Australia', 'Lifeexp'] += 2
df
```

```python
# Find all the rows that contain NA values
rows_with_nan = df[df.isnull().any(axis=1)]
rows_with_nan
```

```python
# Fill NA values with 0 for numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
df
```
