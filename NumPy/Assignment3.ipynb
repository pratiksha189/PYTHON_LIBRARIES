```python
# Normalize a NumPy array [ to normalize means scaling the values of an array into a specific range, often [0, 1]]
```

```python
arr1 = np.array([10,20,30,40,50])
norm = (arr1 - arr1.min()) / (arr1.max() - arr1.min())
print(norm)
```

```python
# Broadcast a 1D array over a 2D array [all possible scenarios]
```

```python
a=np.array([[8,21,4,5,4],[3,4,5,6,7]])
b=np.array([9,4,5,2,7])
num=a+b
num
```

```python
# Sort a 2D array by column 1 [hint: check method argsort()]
```

```python
num = np.array([[10,  5, 39, 89, 73],
                [96,  4, 81, 27,  6],
                [33, 49, 91, 14, 87],
                [29, 68, 57, 70, 79],
                [47, 10,  1, 41, 16]
               ])
sorted_arr = num[num[:,1].argsort()]
sorted_arr
```

```python
# Find common elements between two arrays [with and without using intersect1d, use isin() instead]
```

```python
a=np.array([8,21,4,5,4])
b=np.array([9,4,5,2,7])
common = np.intersect1d(a,b)
common
```

```python
# Subtract row mean from each row of a 2D array
```

```python
arr1 = np.array([[5,2,8],[9,5,3]])
mean_arr = arr1.mean(axis=1, keepdims=True)
final = arr1-mean_arr
final
```

```python
# Given a 1D array of random numbers, find the indices of the 5 largest values (no sorting/ check argsort() instead).
```

```python
unsorted_arr = np.random.randint(4,89,10)
unsorted_arr
```

```python
top_5_indices = np.argsort(unsorted_arr)[-5:][::-1]
unsorted_arr[top_5_indices]
```
